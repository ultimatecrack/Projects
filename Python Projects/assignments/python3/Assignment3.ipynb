```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class Operations:
    def show(self, content):
        print(content)
   
    def purchase(self, rent_time_master, available_vehicles, rate_master):
        self.rate_master = rate_master
        print(rate_master)
        rent_time = input('Time period type: ')
        if rent_time not in rent_time_master:
            print('Not available for {0}'.format(rent_time))
            return

        print(available_vehicles)
        choose_type_of_vehicle = input('Choose vehicle type: ')
        print('Enter vehicle ids and type -1 to finish')
        vehicles = {}
        vehicles[choose_type_of_vehicle] = []
        while True:
            temp_vehicle = input()
            if temp_vehicle == 'bike':
                choose_type_of_vehicle = temp_vehicle
                vehicles[choose_type_of_vehicle] = []
                continue
            elif temp_vehicle == 'cycle':
                choose_type_of_vehicle = temp_vehicle
                vehicles[choose_type_of_vehicle] = []
                continue
            elif temp_vehicle == 'car':
                choose_type_of_vehicle = temp_vehicle
                vehicles[choose_type_of_vehicle] = []
                continue
            else:
                temp_vehicle = int(temp_vehicle)

            #Breaking the loop
            if temp_vehicle == -1:
                break

            #Checking availability
            try:
                available_vehicles[choose_type_of_vehicle][temp_vehicle]
            except:
                print('It is not available. Please check other vehicles')
                continue


            vehicles[choose_type_of_vehicle].append(temp_vehicle)
        purchase_details = {}
        purchase_details['vehicles'] = vehicles
        purchase_details['rent_type'] = rent_time
        return purchase_details
    
    def return_vehicle(self, vehicles,available_vehicles, rent_type, hours = 3):
        #Bill generation
        bill = ''
        total_amount = 0
        for vehicle in vehicles:
            vehicle_details = vehicles[vehicle]
            sub_total = 0
            counter = 0
            for vehicle_id in vehicle_details:
                vehicle_name = available_vehicles[vehicle][vehicle_id]
                rate = self.rate_master[vehicle][rent_type]
                cost = rate * hours
                sub_total += int(cost)
                statement = '{0} ${1}'.format(vehicle_name, cost)
                bill += '\n'+statement
                counter += 1
            print('Cost for {0} is {1}'.format(vehicle, sub_total))
            if (counter >= 3 and counter <= 5):
                print('Congratulation, you unlocked discount for {0}'.format(vehicle))
                if vehicle == 'bike':
                    sub_total = sub_total*0.7
                if vehicle == 'car':
                    sub_total = sub_total*0.6
                if vehicle == 'cycle':
                    sub_total = sub_total*0.5
            total_amount += sub_total
        bill += '\nTotal: ${0}'.format(total_amount)
        print(bill)
```

```python
class VehicleStore:
    def __init__(self):
        #Types of Vehicles Master
        self.vehicle_master = ['bike', 'car', 'cycle']
        
        #Rent Time Master
        self.rent_time_master = ['hourly', 'daily', 'weekly']
        
        #Rate Master
        self.rate_master = {}
        self.rate_master['cycle'] = {'hourly':2, 'daily':8, 'weekly':30, 'discount':50}
        self.rate_master['bike'] = {'hourly':5, 'daily':20, 'weekly':60, 'discount':30}
        self.rate_master['car'] = {'hourly':10, 'daily':40, 'weekly':120, 'discount':40}
        self.rate_master = pd.DataFrame(self.rate_master)
        
        #Available Vehicles
        self.available_vehicles = {}
        self.available_vehicles['cycle'] = ['Hero', 'Racing', 'Lady Bird', 'Rally', 'Hercules']
        self.available_vehicles['bike'] = ['Hero', 'Honda', 'Royal Enfield', 'ZMR', 'Yamaha']
        self.available_vehicles['car'] = ['Audi', 'Maruti Suzuki', 'BMW', 'Mercedez']
        
        print('1. Types vehicles')
        print('2. Display available vehicles')
        print('3. Price List')
        print('4. To Purchase')
        print('5. To return')
        
        customer_input = int(input())
        op = Operations()
        if customer_input == 1:
            op.show(self.vehicle_master)
        elif customer_input == 2:
            op.show(self.available_vehicles)
        elif customer_input == 3:
            op.show(self.rate_master)
        elif customer_input == 4:
            purchased_vehicles = op.purchase(self.rent_time_master, self.available_vehicles, self.rate_master)
            vehicles = purchased_vehicles['vehicles']
            rent_type = purchased_vehicles['rent_type']
            available_vehicles = self.available_vehicles
            print('You have choosen')
            #Bill generation
            bill = ''
            total_amount = 0
            for vehicle in vehicles:
                vehicle_details = vehicles[vehicle]
                for vehicle_id in vehicle_details:
                    vehicle_name = available_vehicles[vehicle][vehicle_id]
                    rate = self.rate_master[vehicle][rent_type]
                    statement = '{0} ${1}'.format(vehicle_name, rate)
                    bill += '\n'+statement
            print(bill)
        elif customer_input == 5:
            purchased_vehicles = op.purchase(self.rent_time_master, self.available_vehicles, self.rate_master)
            time_period = int(input('Total Time: '))
            print('Invoice:')
            if(purchased_vehicles is None):
                return
            op.return_vehicle(purchased_vehicles['vehicles'],self.available_vehicles, purchased_vehicles['rent_type'], time_period)
```
