```python
# create a Statistics dataframe
stats_df= pd.DataFrame(columns = ["no of present", "lessthan50", "between50and60","between60and70", "between70and80", "greaterthan80"],index=["quiz_1","quiz_2"])

stats_df["no of present"]= [quiz1["Firstname"].count(),quiz2["Firstname"].count()]

stats_df["lessthan50"] = [quiz1[quiz1["Grade/10.00"].between(0,4.9,inclusive=True)]["Firstname"].count(),
                         quiz2[quiz2["Grade/10.00"].between(0,4.9)]["Firstname"].count()]

stats_df["between50and60"] = [quiz1[quiz1["Grade/10.00"].between(5,5.9,inclusive=True)]["Firstname"].count(),
                             quiz2[quiz2["Grade/10.00"].between(5,5.9,inclusive=True)]["Firstname"].count()]

stats_df["between60and70"] = [quiz1[quiz1["Grade/10.00"].between(6,6.9,inclusive=True)]["Firstname"].count(),
                             quiz2[quiz2["Grade/10.00"].between(6,6.9,inclusive=True)]["Firstname"].count()]

stats_df["between70and80"] = [quiz1[quiz1["Grade/10.00"].between(7,7.9,inclusive=True)]["Firstname"].count(),
                             quiz2[quiz2["Grade/10.00"].between(7,7.9,inclusive=True)]["Firstname"].count()]

stats_df["greaterthan80"] = [quiz1[quiz1["Grade/10.00"].between(8,10,inclusive=True)]["Firstname"].count(),
                            quiz2[quiz2["Grade/10.00"].between(8,10,inclusive=True)]["Firstname"].count()]

stats_df
```

```python
def get_batch_details(cond, batchdata, quiz):
    filter1 = None
    filter2 = None
    quiz_merged_dataset = pd.merge(batchdata, quiz, how='inner', left_on='studentName', right_on='Firstname')
    if cond == 'noofpresent':
        filter1 = quiz_merged_dataset['Grade/10.00'] == -1
    elif cond == 'lessthan50':
        filter1 = quiz_merged_dataset['Grade/10.00'] < 5
    elif cond == 'between50and60':
        filter1 = quiz_merged_dataset['Grade/10.00']  >= 5
        filter2 = quiz_merged_dataset['Grade/10.00'] <= 6
    elif cond == 'between60and70':
        filter1 = quiz_merged_dataset['Grade/10.00']  >= 6
        filter2 = quiz_merged_dataset['Grade/10.00'] <= 7
    elif cond == 'between70and80':
        filter1 = quiz_merged_dataset['Grade/10.00']  >= 7
        filter2 = quiz_merged_dataset['Grade/10.00'] <= 8
    elif cond == 'greaterthan80':
        filter1 = quiz_merged_dataset['Grade/10.00']  > 8
        
    if filter2 is None:
        combined_condition = filter1
    else:
        combined_condition = filter1 & filter2
        
    quiz_data = str(quiz_merged_dataset.where(combined_condition)['Grade/10.00'].count())
    
    return quiz_data
```

```python
#Importing batches to dictionary
batchwiselist = {}
for file in os.listdir('batchwiselist'):
    temp_name = int(os.path.splitext(file)[0])
    temp_quiz = pd.read_csv(os.path.join('batchwiselist', file))
    batchwiselist[temp_name] = temp_quiz
```

```python
test_cases = {}
number_of_test_cases = int(input('Number of test cases: '))
for i in range(number_of_test_cases):
    batchfile_name = input('Enter batch file name: ')
    condition_to_check = input('Enter condition to check: ')
    test_cases[batchfile_name] = condition_to_check
```

```python
test_cases
```

```python
counter = 0
for test in test_cases:
    temp_batch_file_name = int(test.split(sep='.')[0])
    batchdata = batchwiselist[temp_batch_file_name]
    condition = test_cases[test]
    quiz1_output = get_batch_details(condition, batchdata, quiz1)
    quiz2_output = get_batch_details(condition, batchdata, quiz2)
    filename = 'output{0}.txt'.format(counter)
    with open(filename, 'w') as file:
        file.write(quiz1_output+'\n'+quiz2_output)
    counter += 1
```
